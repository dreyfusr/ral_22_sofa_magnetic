```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# load data
csv_sim = '21_rod_sim_tip_topic_9976/tip_pose_9976_2022-02-15-17-31-18.csv'
csv_track = '11_video_tracker_9976_2/20220215_MVI_9976_trim_1.csv'

df_sim = pd.read_csv(csv_sim)
df_track = pd.read_csv(csv_track)

# parse data from simulation
time_sim = df_sim['%time'].to_numpy()
tip_pos_sim = np.empty([3,len(df_sim)])
tip_pos_sim[0] = df_sim['field.position.x'].to_numpy()
tip_pos_sim[1] = df_sim['field.position.y'].to_numpy()+0.002

# parse data from tracking
time_track = df_track['t'].to_numpy()
tip_pos_track = np.empty([3,len(df_track)])
tip_pos_track[0] = df_track['x'].to_numpy()-0.001
tip_pos_track[1] = df_track['y'].to_numpy()+0.004
```

```python
# plot
fig_1, axs = plt.subplots(1, 1)

axs.plot(tip_pos_sim[0], tip_pos_sim[1], '-', linewidth=2, label="sim")
axs.plot(tip_pos_track[0], tip_pos_track[1], '-', linewidth=2, label="exp")

axs.set_aspect('equal', 'box')
axs.set_xlabel('x (m)', fontsize=16)
axs.set_ylabel('y (m)', fontsize=16)
handles, labels = axs.get_legend_handles_labels()
axs.legend(loc='lower left', fontsize=12, frameon=False)
axs.tick_params(axis='both', which='major', labelsize=12)
axs.set_xlim([-0.1, 0.1])
axs.set_ylim([-0.1, 0.1])
axs.locator_params(nbins=3)
axs.tick_params(right=True, top=True)
```

```python
from scipy import signal
# resampling data

num_samples = 500

tip_pos_sim[0]
tip_pos_sim[1]
x_sim = signal.resample(tip_pos_sim[0], num_samples)
y_sim = signal.resample(tip_pos_sim[1], num_samples)

tip_pos_track[0]
tip_pos_track[1]
x_track = signal.resample(tip_pos_track[0], num_samples)
y_track = signal.resample(tip_pos_track[1], num_samples)

time_common = np.linspace(0, 10, num_samples, endpoint=True)

# plot
fig_2, axs = plt.subplots(1, 1)

axs.plot(x_sim*1e3, y_sim*1e3, ':', linewidth=2, label="sim")
axs.plot(x_track*1e3, y_track*1e3, ':', linewidth=2, label="exp")

axs.set_aspect('equal', 'box')
axs.set_xlabel('x (mm)', fontsize=16)
axs.set_ylabel('y (mm)', fontsize=16)
handles, labels = axs.get_legend_handles_labels()
axs.legend(loc='lower left', fontsize=12, frameon=False)
axs.tick_params(axis='both', which='major', labelsize=12)
axs.set_xlim([-100, 100])
axs.set_ylim([-100, 100])
axs.locator_params(nbins=3)
axs.tick_params(right=True, top=True)
```

```python
# position error between simulation and experiment

pos_error = np.sqrt((x_sim-x_track)**2 + (y_sim-y_track)**2)

# plot
fig_3, axs = plt.subplots(1, 1)

axs.plot(time_common*1e3, pos_error*1e3, '-', linewidth=2, label="sim")

#axs.set_aspect(1.0/axs.get_data_ratio(), adjustable='box')
axs.set_xlabel('time (s)', fontsize=16)
axs.set_ylabel('error (mm)', fontsize=16)
handles, labels = axs.get_legend_handles_labels()
axs.tick_params(axis='both', which='major', labelsize=12)
#axs.set_xlim([-0.1, 0.1])
#axs.set_ylim([0, 70])
axs.locator_params(nbins=3)
```

```python
# export figure

# make text editable in inkskape
mpl.use('svg')
new_rc_params = {'text.usetex': False,
"svg.fonttype": 'none'
}
mpl.rcParams.update(new_rc_params)

# export
fig_1.savefig("plot_vessels.svg", transparent=True)
```
